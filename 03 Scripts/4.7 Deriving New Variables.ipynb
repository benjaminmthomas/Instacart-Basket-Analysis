```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Busiest days

result = []

for value in df_ords_prods_merge["order_dow"]:
  if value == 0:
    result.append("Busiest days")
  elif value == 1:
    result.append("Busiest days")    
  elif value == 4:
    result.append("Slowest days")
  elif value == 3:
    result.append("Slowest days")    
  else:
    result.append("Regularly busy")
```

```python
# New column for busiest hours of the day by splitting in three groups of 8 hours each

result = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result.append("Most orders")
  elif value in [3, 4, 2, 5, 1, 0, 6, 23]:
    result.append("Fewest orders")    
  else:
    result.append("Average orders")
```
